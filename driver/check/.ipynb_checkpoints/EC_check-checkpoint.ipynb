```python
import xarray
import pygmt
import pandas
import numpy as np
from matplotlib.figure import Figure
from matplotlib.colors import ListedColormap,LinearSegmentedColormap
from matplotlib.ticker import MultipleLocator
from IPython.display import display
```

```python
Nlvdplr = 80
Nlvtemp = 57
NlvdBZe = 35

window = 0.2
```

```python
temp = -79.0 + 2.0 * np.arange(Nlvtemp)
zgrd = -39.0 + 2.0 * np.arange(NlvdBZe)
vgrd =  -7.9 + 0.2 * np.arange(Nlvdplr)
```

```python
data = np.fromfile("../data/EarthCARE/cfed_Vd.bin", dtype=np.float32)
data = data.reshape( (5,5,Nlvtemp,Nlvdplr) )
```

```python
# TR:0 NM:1 NP:2 SM:3 SP:4
region=1
```

```python
# warm:0 upper:1 conv:2 strat:3 other:4 all:5 w/o warm:6
ctype=2
```

```python
if ctype<=4:
    var = data[region,ctype,:,:]
elif ctype==5:
    var = data[region,1:4,:,:]
elif ctype==6:
    var = data[region,2:4,:,:]
```

```python
smpl = ( var.sum(axis=1) * window )[:,None]
nrm = np.broadcast_to(smpl,var.shape)
cfed = var/nrm
```
