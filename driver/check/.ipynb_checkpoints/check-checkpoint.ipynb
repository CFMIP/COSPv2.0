```python
import xarray
import pygmt
import pandas
import numpy as np
```

```python
ds = xarray.open_dataset("../data/outputs/UKMO/cosp2_output.um_global.gfortran.kgo.v006.nc")
```

```python
print(ds.info())
```

```python
print(ds["lev"].values)
```

```python
lat = ds["latitude"].values.ravel()
cl  = ds["cltatlid"].values.ravel()
```

```python
mask = np.isfinite(cl) & (cl >= 0.) & (cl <= 100.)
lon, lat, cl = lon[mask], lat[mask], cl[mask]
```

```python
print(cl.size)
```

```python
spacing = "10/10"
region = [0, 360, -90, 90]
```

```python
grid = pygmt.xyz2grd(
    data = np.column_stack([lon, lat, cl]),
    region = region,
    spacing = spacing,
)
```

```python
pygmt.makecpt(
    cmap="viridis",
    series=[0.0, 100.0, 1.0]
)
```

```python
fig = pygmt.Figure()
```

```python
fig.coast(
    region="g",                # global
    projection="W0/15c",       # 投影（例）
    resolution="c",
    land="white",              # 陸を白で塗る
    water="white",             # 海を白で塗る（"transparent" は NG -> "white" を使用）
    frame=["af", "+t'CALIPSO Ice Total Cloud Fraction (%)'"],
)
```

```python
fig.grdimage(
    grid = grid,
    projection = "W0/15c",
    region = "g",
    frame = False,
)
```

```python
fig.coast(shorelines="1/0.5p", resolution="c")

fig.colorbar(frame=["a", "x+lm"])
```

```python
fig.show(width="80%")
```
